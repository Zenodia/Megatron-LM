```python
import os , sys
import numpy as np
import pandas as pd

sts_path='/workspace/SVdata/SuperLimSE/sts/'

train_df=pd.read_csv(sts_path+'train-sv.tsv', sep='\t')
test_df=pd.read_csv(sts_path+'test-sv.tsv', sep='\t')
dev_df=pd.read_csv(sts_path+'dev-sv.tsv', sep='\t')
train_df.head()
```

```python
print("train records count : ", len(train_df))
print("test records count : ", len(test_df))
print("dev records count : ", len(dev_df))
```

```python
def round2class(x):    
    return int(round(float(x)))

def shrink2three(x):
    x_=round2class(x)
    if x_==0:
        return 0
    elif 0<x_<=3:
        return 1
    elif 3<x_<=5:
        return 2
    else:
        print("x_:{} is not in [0,1,2]".format(str(x_)))

train_df['label']=train_df['score'].apply(lambda x : round2class(x))
test_df['label']=test_df['score'].apply(lambda x : round2class(x))
dev_df['label']=dev_df['score'].apply(lambda x : round2class(x))
train_df.head()
```

```python
train_df['index']=train_df.index
cur_index=train_df.index[-1]+1
n=len(test_df)
test_df['index']=[cur_index+i for i in range(n)]
test_df['index'].values[-1]
cur_index=test_df['index'].values[-1]+1
n=len(dev_df)
dev_df['index']=[cur_index+i for i in range(n)]
```
